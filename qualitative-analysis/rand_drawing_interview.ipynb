```python

share = 0.08
conv_folder = "../single-factor-experiments/bagel-dpo-34b-v0.2-bagel-dpo-34b-v0.2/"
conv_folder = "../single-factor-experiments/bagel-dpo-34b-v0.2-Yi-34B-Chat/"
conv_folder = "../single-factor-experiments/Yi-34B-Chat-bagel-dpo-34b-v0.2/"
conv_folder = "../single-factor-experiments/Yi-34B-Chat-Yi-34B-Chat/"

rand_conv_folder = "bagel-dpo-34b-v0.2-bagel-dpo-34b-v0.2_interview/"
rand_conv_folder = "bagel-dpo-34b-v0.2-Yi-34B-Chat_interview/"
rand_conv_folder = "Yi-34B-Chat-bagel-dpo-34b-v0.2_interview/"
rand_conv_folder = "Yi-34B-Chat-Yi-34B-Chat_interview/"

interview_folder = "../COMPLETE_bagel-dpo-34b-v0.2-bagel-dpo-34b-v0.2/"
interview_folder = "../COMPLETE_bagel-dpo-34b-v0.2-Yi-34B-Chat/"
interview_folder = "../COMPLETE_Yi-34B-Chat-bagel-dpo-34b-v0.2/"
interview_folder = "../COMPLETE_Yi-34B-Chat-Yi-34B-Chat/"

for i in os.listdir(conv_folder):
    no_draws = round(len(os.listdir(conv_folder + i))*share)
    print(no_draws)
    draw_conv = random.sample(range(0, len(os.listdir(conv_folder + i))), no_draws)
    print(draw_conv)
    if not os.path.exists(rand_conv_folder + i):
        os.makedirs(rand_conv_folder + i)
    else:
       shutil.rmtree(rand_conv_folder + i)           # Removes all the subdirectories!
       os.makedirs(rand_conv_folder + i)
    for file in draw_conv:
        # construct full file path
        source = conv_folder + i + "/" + os.listdir(conv_folder+i)[file]
        destination = rand_conv_folder + i + "/" + os.listdir(conv_folder+i)[file]
        # move file
        shutil.copytree(source, destination,dirs_exist_ok=True)
        print('Copied:', file)
    
    
    
    
    
    
```

```python
rows_list =[]

for i in os.listdir(rand_conv_folder):
    name = [i] * len(os.listdir(rand_conv_folder + i))
    identifier = os.listdir(rand_conv_folder + i)
    label = ["nan"] * len(os.listdir(rand_conv_folder + i))
    text = []
    for item in os.listdir(rand_conv_folder + i):
        with open(rand_conv_folder + i + "/" + item + "/" +"conversation.json") as f:
            d = json.load(f)
            text.append(d)
    rows_list.append([name, identifier, text, label])
    

    
print(rows_list)
df_to_label = pd.DataFrame(rows_list, columns=["name", "identifier", "text", "label"]).reset_index()
df_to_label = df_to_label.explode(["name", "identifier", "text", "label"])

selected_conversations = list(df_to_label["identifier"])
len(set(selected_conversations)) # 38 unique observations
```

```python
# append interviews: look for identifier and append the two interviews
interviews_list =[]

for i in os.listdir(interview_folder):
    name_folder = os.path.join(interview_folder, i)
    for identificator in os.listdir(name_folder):
        identificator_folder = os.path.join(name_folder, identificator)
        identifier_interview = identificator
        name_interview = i        
        text = []
        #for item in os.listdir(identificator_folder):
        for person in ["renter", "landlord"]:
            try:
                with open(identificator_folder  + "/" + person + "/" + "conversation.json") as f:
                    d = json.load(f)
                    text.append(d)
                interviews_list.append([name_interview, identifier_interview, text, person])
                print("successfully appended: ", identificator_folder  + "/" + person + "/" + "conversation.json")
            except Exception as e:
                print(f"An error occured: {e}. Trying again.")
                interviews_list.append([name_interview, identifier_interview, "nan", person])

df_interviews = pd.DataFrame(interviews_list, columns=["name_interview", "identifier_interview", "text", "person"]).reset_index()
df_interviews_renter = df_interviews[df_interviews["person"] == "renter"]
df_interviews_renter.rename(columns={'text':'interview_renter'}, inplace=True)
df_interviews_landlord = df_interviews[df_interviews["person"] == "landlord"]
df_interviews_landlord.rename(columns={'text':'interview_landlord'}, inplace=True)
#df_interviews = df_to_label.explode(["name", "identifier", "text", "person"])
print("Length of interview df: ", len(df_interviews))
print("Length of renter interview df: ", len(df_interviews_renter))
print("Length of landlord interview df: ", len(df_interviews_landlord))
df_interviews_landlord
```

```python
# merge interview and conversations to one df
print("Length of Interview df: ", len(df_interviews))
print("Length of Conversation df: ", len(df_to_label))
full_to_label = df_to_label.merge(df_interviews_renter, how="left", left_on=["name", "identifier"], right_on= ["name_interview","identifier_interview"])
full_to_label = full_to_label.merge(df_interviews_landlord, how="left", left_on=["name", "identifier"], right_on= ["name_interview","identifier_interview"], indicator=True)
print("Length of full df: ", len(full_to_label))
display(full_to_label)
```
